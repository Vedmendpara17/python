## Typecasting in Python :-

#### FOR int() :-

```python
# when we take input from user then by default it will store in variable as a string 
# at that point typecasting topic is comming in the picture.
# For float value.
a = 2.3
print(type(a))
num = int(a)
print(num,type(num))
```

```python
# for boolean (bool)
a = True
num = int(a)
print(num,type(num))

b = False
nums = int(b)
print(nums,type(nums))
```

```python
# for string
# string to integer conversion possible but not for (a,b,c,....) and (1.2,....)
a = "12"
nums = int(a)
print(nums,type(nums))
```

#### FOR float() :-

```python
# for float
a = 152
num = float(a)
print(num,type(num))
```

```python
# for boolean(bool)
a = True
num = float(a)
print(num,type(num))
```

```python
# for string
a = "11222"
num = float(a)
print(num,type(num))

b = "12.3"
nums = float(b)
print(nums,type(nums))

c = "12.3.2"  
numc = float(c) # this giving error because double point is not allow by python.
print(numc,type(numc))
```

#### FOR bool() :-

```python
# for int
b = 1
num = bool(b)
print(num,type(num)) # it will give "True <class 'bool'>"

c = 1245.45
numc = bool(c)
print(numc,type(numc))
```

```python
# for float
c = 0.001
num = bool(c)
print(num,type(num))
```

```python
# for string
k = "1.2222"
num = bool(k)
print(num,type(num))

z = "kishan"
nums = bool(z)
print(nums,type(nums))

kis = ""
numc = bool(kis)
print(numc,type(numc))

kic = "1.2.3"
numk = bool(kic)
print(numk,type(numk))
```

### Membership Operators in Python :-

```python
# for " in " operator
print("H" in "python") # it will give False because python is case sencitive
# we can also check substring present in word.
print("thon" in "python")
```

```python
# for " not in "
print("h" not in "python") # it will give False because 'not' keyword is reverse result of in
```

# Precedence :- (Higher to Lower)
1. **   -> "Right to Left"
2. /,*   -> "Left to Right"
3. +,-   -> "Left to Right"
4. comparision operators   -> "Left to Right"
5. not   -> "Left to Right"
6. and   -> "Left to Right"
7. or   -> "Left to Right"

```python
# for power
2**3**1 # output 8
# using () for '**'
((2**3)**2) # output 64
```

```python
a = 6
b = 0
c = 'py'
print((a or b) and (b or c) and ('Hello'))
print((a or b) and (b or c) and not ('Hello'))
```
